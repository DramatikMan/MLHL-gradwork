```python
import random
from matplotlib import pyplot as plt
from pathlib import Path

images: list[Path] = []
train_path = Path("data/train")

for image_class in train_path.iterdir():
    image_path = random.choice(sorted(image_class.iterdir()))
    images.append(image_path)

fig = plt.figure(1, figsize=(10, 15))

for subplot, image in enumerate(images):
    image_class = image.parent.name
    image = plt.imread(image)
    fig = plt.subplot(5, 3, subplot + 1)
    fig.set_title(image_class, pad=10, size=18)
    fig.set_xticklabels([])
    fig.set_yticklabels([])
    plt.imshow(image)

plt.tight_layout()
```

```python
corner_cases = (
    Path("data/train/Cauliflower/0614.jpg"),
    Path("data/train/Cucumber/0932.jpg"),
    Path("data/train/Pumpkin/0011.jpg"),
    Path("data/train/Papaya/0020.jpg"),
)

fig = plt.figure(2, figsize=(7, 5))

for subplot, image in enumerate(corner_cases):
    image_class = image.parent.name
    image = plt.imread(image)
    fig = plt.subplot(2, 2, subplot + 1)
    fig.set_title(f"true class = {image_class}", pad=10, size=18)
    fig.set_xticklabels([])
    fig.set_yticklabels([])
    plt.imshow(image)

plt.tight_layout()
```
